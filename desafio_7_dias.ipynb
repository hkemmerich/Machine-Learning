```python
def missing_visu(dataset):
    quant_null = dataset.isnull().sum()
    columns = dataset.columns
    dic = {'col':[],'quant_null': []}
    for column,quant in zip(columns,quant_null):
        if quant >0:
            dic['col'].append(quant)
            dic['quant_null'].append(column)
    dataset = pd.DataFrame(dic)
    plt.figure(figsize=(15,5))
    sns.barplot(x=dataset['quant_null'],y=dataset['col'],data=dataset,palette='rocket')
    plt.xticks(rotation=45)
```

```python
n_estimators = [3,5,10,100]

max_features = [3,4,5,6, None]

max_depth = [3,4,5,6,None]

class_weight = ['balance', None]

random_grid = {'n_estimators': n_estimators, 'max_features': max_features, 'max_depth': max_depth, 'class_weight': class_weight}

pprint(random_grid)
```

```python
import numpy as np
```

```python
accuracy_list_rf = []
precision_list_rf = []
reacall_list_rf = []
f1_list_rf = []
auc_list_rf = []


grid_rf = RandomizedSearchCV(RandomForestClassifier(), random_grid, random_state=42)

start = time.time()
count = 0

for train, test in stratifkfold.split(X_train_array, y_train_array):
    
    pipeline_rf = imbalan_make_pipeline(RandomOverSampler(sampling_strategy='minority'), grid_rf)
    
    model_rf = pipeline_rf.fit(X_train_array[train], y_train_array[train])
    
    best_rf = grid_rf.best_estimator_
    
    prediction_rf = best_rf.predict(X_train_array[test])
    
    accuracy_list_rf.append(pipeline_rf.score(X_train_array[test], y_train_array[test]))
    precision_list_rf.append(precision_score(y_train_array[test], prediction_rf))
    reacall_list_rf.append(recall_score(y_train_array[test], prediction_rf))
    f1_list_rf.append(f1_score(y_train_array[test], prediction_rf))
    auc_list_rf.append(roc_auc_score(y_train_array[test], prediction_rf))
    
    count = count +1
    print(f'kfold n: {count}')
    
    intervalo = time.time()
    print(f'tempo intermedi√°rio: {(intervalo - start)/60:.1f} minutos')
    
print('---'*20)
print('Result')
print(f'accuracy: {np.mean(accuracy_list_rf)}')
print(f'prediction: {np.mean(prediction_rf)}')
print(f'recall: {np.mean(reacall_list_rf)}')
print(f'f1 {np.mean(f1_list_rf)}')
print(f'auc: {np.mean(auc_list_rf)}')
print('---'*20)


end = time.time()

print(f'tempo do loop para valida√ß√£o cruzada: {(end-start)/60:.1f} minutos.')
```

```python
def draw_roc_curve(classif, cv, X, y, title='ROC Curve'):
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0,1,100)
    
    i=0
    
    for train, test in cv.split(X, y):
        prob = classif.fit(X.iloc[train],y.iloc[train]).predict_proba(X.iloc[test])
        fpr, tpr, thresholds = roc_curve(y.iloc[test], prob[:, 1])
        tprs.append(interp(mean_fpr, fpr, tpr))
        tprs[-1][0] = 0.0
        roc_auc = auc(fpr, tpr)
        aucs.append(roc_auc)
        
        plt.plot(fpr,tpr, lw=1, alpha=0.3, label='ROC fold %d (AUC = %0.2f)'% (i,roc_auc))
        
        i +=1
    plt.plot([0,1],[0,1], linestyle='--',lw=2,color='r', label='Luck', alpha=.8)
    
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    
    plt.plot(mean_fpr, mean_tpr, color='b', label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' %(mean_auc, std_auc), lw=2, alpha=.8)
    
    std_tpr = np.std(tprs, axis=0)
    tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
    tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
    plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=0.2, label=r'$\pm$ 1 std. dev.')
    
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False positive rate', fontsize=12)
    plt.ylabel('True positive rate', fontsize=12)
    plt.title(title)
    plt.legend(loc='lower right')
    
    plt.show()
```

```python
def remove_corr_var(data, value_cut = 0.95):
    matrix_corr = data.iloc[:, 4:-2].corr().abs()
    matrix_upper = matrix_corr.where(np.triu(np.one(matrix_corr.shape),k=1).astype(np.bool))
    delete = [column for column in matrix_upper.columns if any(matrix_upper[column] > value_cut)]
    
    return data.drop(delete, axis=1)
```
